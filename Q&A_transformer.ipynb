```python
from IPython import get_ipython
from IPython.display import display
# %%
#Instead of importing from 'transforms', we will use 'transformers' for Bert
from transformers import BertForQuestionAnswering, BertTokenizer
import torch

model_name="deepset/bert-base-cased-squad2"
tokenizer=BertTokenizer.from_pretrained(model_name)
model=BertForQuestionAnswering.from_pretrained(model_name)
```

```python
context="""The Eiffel Tower is a wrought-iro lattice tower on the camp de mars in paris,france.
It is named after the enginer Gustave Eiffel,whose company designed and built the tower."""
question="Who designed the Eiffel Tower?"

input=tokenizer(question,context,return_tensors="pt")
with torch.no_grad():
  output=model(**input)
  start_scores=output.start_logits
  end_scores=output.end_logits
```

```python
start_idx=torch.argmax(start_scores)
end_idx=torch.argmax(end_scores)+1
answer=tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input["input_ids"][0][start_idx:end_idx]))
print(answer)
```
